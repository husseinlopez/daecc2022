### Datos sint√©ticos: medias similares

```r
x <- rnorm(10, mean = 100, sd = 10)
y <- rnorm(10, mean = 100, sd = 10)
data <- data.frame(x,y)
boxplot(data)
```

```r
shapiro.test(x)
shapiro.test(y)
```

### Datos sint√©ticos: medias distintas, n>30 y diferentes longitudes

```r
x <- rnorm(100, mean = 100, sd = 10)
y <- rnorm(500, mean = 200, sd = 10)
data <- data.frame(x,y)
boxplot(data)
```

```r
shapiro.test(x)
shapiro.test(y)
```

#### two.sided: ùúá = ùúá0

```r
t.test(x,y)
# two.sided: H0: ùúá = ùúá0 , H1: ùúá ‚â† ùúá0
# Si p > 0.05, no se rechaza la hip√≥tesis que las medias son iguales
```

#### one.sided: ùúá ‚â§ ùúá0

```r
t.test(x,y,alternative="greater")
# one.sided: H0: ùúá ‚â§ ùúá0, H1: ùúáÔπ•ùúá0
# Si p > 0.05, no se rechaza la hip√≥tesis nula: la media de x es menor que la media de y
```

#### one.sided: ùúá ‚â• ùúá0

```r
t.test(x,y,alternative="less")
# one.sided: H0: ùúá ‚â• ùúá0, H1: ùúáÔπ§ùúá0
# Si p < 0.05, se rechaza la hip√≥tesis nula, por lo tanto no se rechaza la alternativa.
```

### Ejemplo:

Se tienen dos muestras de las notas del examen final de dos cursos:

Curso A (te√≥rico-pr√°ctico): 15, 12, 11, 18, 15, 15,9, 19, 14, 13, 11, 12, 18, 15, 16, 14, 16,17, 15, 17, 13, 14, 13, 15, 17, 19, 17, 18, 16 y 14.

Curso B (te√≥rico): 11, 16, 14, 18,6,8,9, 14, 12, 12, 10, 15, 12,9, 13, 16, 17, 12,8,7, 15,5, 14, 13, 13, 12, 11, 13, 11 y 7.

El examen fue id√©ntico para ambos grupos y se pod√≠a obtener un m√°ximo de veinte.

Se desea saber, con un nivel de significancia de 0.05, si existe diferencia entre la media de ambos grupos.

```r
Grupo.A = c(15, 12, 11, 18, 15, 15, 9, 19, 14, 13, 11, 12, 18, 15, 16, 14, 16, 17, 15, 17, 13, 14, 13, 15, 17, 19, 17, 18, 16, 14)
Grupo.B = c(11, 16, 14, 18, 6, 8, 9, 14, 12, 12, 10, 15, 12, 9, 13, 16, 17, 12, 8, 7, 15, 5, 14, 13, 13, 12, 11, 13, 11, 7)
data <- data.frame(Grupo.A, Grupo.B)
boxplot(data)
```

```r
shapiro.test(Grupo.A)
shapiro.test(Grupo.B)
```

```r
t.test(Grupo.A,Grupo.B)
# two.sided: H0: ùúá = ùúá0, H1: ùúá ‚â† ùúá0
# Si p < 0.05, se rechaza la hip√≥tesis nula, por lo tanto hay una diferencia significativa entre las notas de ambos grupos
```

```r
t.test(Grupo.A,Grupo.B, alternative = "g")
# one.sided: H0: ùúá ‚â§ ùúá0, H1: ùúáÔπ•ùúá0
# Si p < 0.05, se rechaza la hip√≥tesis nula, por lo tanto la media del grupo A es mayor a la media del grupo B
```

```r
t.test(Grupo.A,Grupo.B, alternative = "l")
# one.sided: H0: ùúá ‚â• ùúá0, H1: ùúáÔπ§ùúá0
# Si p > 0.05, no se rechaza la hip√≥tesis nula, por lo tanto la media del grupo A es mayor o igual a la del grupo B
```

## 2. Prueba para datos param√©tricos con grupos relacionados (pareados)

### Datos sint√©ticos: medias distintas y n>30

```r
x <- rnorm(100, mean = 100, sd = 10)
y <- rnorm(100, mean = 200, sd = 10)
data <- data.frame(x,y)
boxplot(data)
```

```r
shapiro.test(x)
shapiro.test(y)
```

```r
t.test(x,y, paired = TRUE)
# two.sided: H0: ùúá = ùúá0, H1: ùúá ‚â† ùúá0
# Si p < 0.05, se rechaza la hip√≥tesis nula, por lo tanto las medias de las variables son distintas
```

### Ejemplo:

Dado un grupo de veinte estudiantes, se quiere investigar el efecto del uso de alg√∫n recurso did√°ctico (con un nivel de significancia de 0.05), en su destreza para resolver cierto tipo de problemas matem√°ticos. 

Se les toma un test inicial, despu√©s ven el recurso, y al finalizar se les toma un test adicional.

(Ver tabla en diapositivas de clase 8)

```r
datos = data.frame(
   Pre = c(18, 21, 16, 22, 19, 24, 17, 21, 23, 18, 14, 16, 16, 19, 18, 20, 12, 22, 15, 17),
   Post = c(22, 25, 17, 24, 16, 29, 20, 23, 19, 20, 15, 15, 18, 26, 18, 24, 18, 25, 19, 16)
)
boxplot(datos)
```

```r
shapiro.test(datos$Pre)
shapiro.test(datos$Post)
```

```r
t.test(datos$Pre,datos$Post, paired = T)
# two.sided: H0: ùúá = ùúá0, H1: ùúá ‚â† ùúá0
# Si p < 0.05, se rechaza la hip√≥tesis nula, por lo tanto hay una diferencia significativa entre las medias de las tomas
```

```r
t.test(datos$Pre,datos$Post, paired = T, alternative = "g")
# one.sided: H0: ùúá ‚â§ ùúá0, H1: ùúáÔπ•ùúá0
# Si p > 0.05, no se rechaza la hip√≥tesis nula, por lo tanto la media del test Pre es menor o igual a la del test Post
```

```r
t.test(datos$Pre,datos$Post, paired = T, alternative = "l")
# one.sided: H0: ùúá ‚â• ùúá0, H1: ùúáÔπ§ùúá0
# Si p < 0.05, se rechaza la hip√≥tesis nula, por lo tanto la media del test Pre es menor a la media del test Post
```

## 3. Prueba para datos no param√©tricos con grupos independientes

### Datos sint√©ticos: medianas distintas y n>30

```r
x <- runif(100, min = 0, max = 100)
y <- runif(100, min = 50, max = 200)
data <- data.frame(x,y)
boxplot(data)
```

```r
shapiro.test(x)
shapiro.test(y)
```

```r
wilcox.test(x,y)
# two.sided: 
#   H0: P(x>y) = P(y>x), H0: P(x>y) = 0.5
#   H1: P(x>y) ‚â† P(y>x), H1: P(x>y) ‚â† 0.5
# Si p < 0.05, se rechaza la hip√≥tesis nula, por lo tanto las medianas de las variables son distintas
```

### Ejemplo:

Suponemos que tenemos datos diagn√≥sticos de cuatro mujeres y cinco hombres. 

Todos fueron diagnosticados con diabetes y tenemos la edad a la cual se les descubri√≥ la enfermedad.

Los datos son:
Hombres: {19, 22, 16, 29, 24},
Mujeres: {20, 11, 17, 12}.

Se desea saber si hay diferencia en la edad entre hombres y mujeres considerando una significancia del 0.05. 

```r
Hombres <- c(19, 22, 16, 29, 24)
Mujeres <- c(20, 11, 17, 12)
boxplot(Hombres, Mujeres)
```

```r
shapiro.test(Hombres)
shapiro.test(Mujeres)
```

```r
wilcox.test(Hombres,Mujeres)
# two.sided: 
#   H0: P(x>y) = P(y>x), H0: P(x>y) = 0.5
#   H1: P(x>y) ‚â† P(y>x), H1: P(x>y) ‚â† 0.5
# Si p < 0.05, se rechaza la hip√≥tesis nula, por lo tanto las medianas de los grupos por g√©nero son distintas
```

### Ejemplo:

Se tiene un grupo de doce pacientes con artritis a los que se les suministra dos medicaciones distintas para aliviar los s√≠ntomas. 

Se les pidi√≥ a todos que indicaran cu√°ntas horas de alivio observaron con ambos medicamentos.

Se desea saber si existe una diferencia significativa de usar uno u otro medicamento.

(Ver tabla en diapositivas de clase 8)

```r
datos = data.frame(
  Med.A = c( 2,  3.6,  2.6,  2.7,  7.3,  3.4,  14.9,  6.6,  2.3,  2.1,  6.8,  8.5),
  Med.B = c( 3.5,  5.7,  2.9,  2.4,  9.9,  3.3,  16.7,  6,  3.8,  4,  9.1,  20.9)
)
boxplot(datos)
```

```r
shapiro.test(datos$Med.A)
shapiro.test(datos$Med.B)
```

```r
wilcox.test(datos$Med.A, datos$Med.B, paired = T)
# two.sided: 
#   H0: P(x>y) = P(y>x), H0: P(x>y) = 0.5
#   H1: P(x>y) ‚â† P(y>x), H1: P(x>y) ‚â† 0.5
# Si p < 0.05, se rechaza la hip√≥tesis nula, por lo tanto hay una diferencia estad√≠sticamente significativa entre los medicamentos
```
