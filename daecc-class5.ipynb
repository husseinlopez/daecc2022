## 4. Pruebas de normalidad

### Histogramas con distribución normal

```r
x <- iris$Sepal.Length #cambiar la variable
plotn <- function(x,main="",
                   xlab="X",ylab="Densidad") {
                   min <- min(x)
                   max <- max(x)
                   media <- mean(x)
                   dt <- sd(x)
                   hist(x,freq=F,main=main,xlab=xlab,ylab=ylab)
                   curve(dnorm(x,media,dt), min, max,add = T,col="blue")
                 }
plotn(x,main="Histograma y distribucion normal iris$Sepal.Length")
```

```r
x <- iris$Sepal.Width
plotn <- function(x,main="",
                   xlab="X",ylab="Densidad") {
                   min <- min(x)
                   max <- max(x)
                   media <- mean(x)
                   dt <- sd(x)
                   hist(x,freq=F,main=main,xlab=xlab,ylab=ylab)
                   curve(dnorm(x,media,dt), min, max,add = T,col="blue")
                 }
plotn(x,main="Histograma y distribucion normal iris$Sepal.Width")
```

```r
x <- iris$Petal.Length 
plotn <- function(x,main="",
                   xlab="X",ylab="Densidad") {
                   min <- min(x)
                   max <- max(x)
                   media <- mean(x)
                   dt <- sd(x)
                   hist(x,freq=F,main=main,xlab=xlab,ylab=ylab)
                   curve(dnorm(x,media,dt), min, max,add = T,col="blue")
                 }
plotn(x,main="Histograma y distribucion normal iris$Petal.Length")
```

```r
x <- iris$Petal.Width 
plotn <- function(x,main="",
                   xlab="X",ylab="Densidad") {
                   min <- min(x)
                   max <- max(x)
                   media <- mean(x)
                   dt <- sd(x)
                   hist(x,freq=F,main=main,xlab=xlab,ylab=ylab)
                   curve(dnorm(x,media,dt), min, max,add = T,col="blue")
                 }
plotn(x,main="Histograma y distribucion normal iris$Petal.Width")
```
