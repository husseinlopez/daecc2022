# Pruebas de asociación

## Chi cuadrado de Pearson (prueba para variables cualitativas)

Ejemplo

En una encuesta de opinión se preguntó a 1230 mujeres y a 961 hombres si están a favor o en contra de una nueva ley.
Se desea saber si el género de la persona está asociado con la opinión.

Mujeres: A favor 762, En contra 468

Hombres: A favor 484, En contra 477

La hipótesis nula es que no hay asociación entre las dos variables, i.e., que el género de la persona no se asocia con su opinión sobre el tema.

```r
datos<- data.frame(
  favor = c( 762, 468),
  contra = c( 484, 477)
)
rownames(datos) <- c("mujeres", "hombres") 
chisq.test(datos)
```

Se compara este valor con el p-value para un grado de libertad, y se rechaza la hipótesis nula concluyendo que el género sí influye en la opinión sobre este tema.

## Coeficiente de Pearson (prueba para variables cuantitativas)

Ejemplo

Resultados de un examen de traducción (x) y comprensión (y) de inglés.

```r
datos<- data.frame(
  x = c( 17, 13, 12, 14, 15, 8, 9, 13, 11, 14, 12, 16),
  y = c( 15, 13, 8, 17, 16, 9, 14, 10, 16, 13, 14, 17)
) 
cor(datos$x, datos$y)
```

## Coeficiente de Spearman (prueba para variables cuantitativas)

Ejemplo

Datos de coeficiente intelectual (X) y horas de tv semanales  (Y).

```r
datos <- data.frame(
  x = c(106, 86, 100, 100, 99, 103, 97, 113, 113, 110),
  y = c(7, 0, 28, 50, 28, 28, 20, 12, 7, 12)
)

cor(datos$x, datos$y, method = "spearman")
```
